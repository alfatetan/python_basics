```python
# Task #6
import re

def clean_element(s: str):
    """
    Leave only digits in the string

    in : str - include different symbols
    out: int - with digits only
    """
    result = re.sub("[^0-9]", "", s)
    return int(result) if result != '' else 0

# Потрошим нужный файл в переменную
with open("file_lesson5_task6.txt", "r") as file:
    data = file.readlines()
    print(data)

# Вычленим название предмета и все цифры, убирая мусор

clean_data = dict() # здесь будет конечный результат вычислений

for line in data:
    dirty_line = line.split() #берём каждую строку со всеми артефактами

    # Считаем количество часов для каждого урока
    lesson_time = 0
    for el in range(1, len(dirty_line)):
        # при вычислении - чистим от ненужных символов с помощью регулярки
        lesson_time += clean_element(dirty_line[el])
    
    clean_data[dirty_line[0]] = lesson_time

print(clean_data)

```

```python
# Task #7

import json

with open("file_lesson5_task7.txt", "r") as file:
    whole_data = file.readlines()

# Тут будет находиться прибыль всех компаний
companies = {}

# Тут будет средняя прибыль по всем компаниям
average_profit = 0

# Количество компаний с прибылью
profit_companies = 0

# Всего прибылей компании, за исключением убытков
total_profit = 0

for line in whole_data:
    #Вычисляем прибыль каждой компании и формируем словарь
    elements = line.split()
    gesheft = int(elements[2]) - int(elements[3])

    # Если компания имеет прибыль, то мы включаем её для расчёта средней прибыли
    if gesheft > 0:
        total_profit += gesheft
        profit_companies += 1

    # формируем отчёт по прибыли/убыткам каждой компании
    companies[elements[0]] = gesheft

# Вычисляем среднюю прибыль компаний с положительным балансом
average_profit = total_profit / profit_companies

#Преобразуем в нужный формат для записи JSON
total_result = [
    companies,
    {"average_profit": average_profit}
]

# Пишем JSON
with open("file_lesson5_task7.json", "w") as file:
    json.dump(total_result, file)
```
